```python
import subprocess
import sys


def check_current_state():
    print("Python Version:", sys.version)

    # Check pip packages
    print("\nInstalled Packages:")
    reqs = subprocess.check_output([sys.executable, "-m", "pip", "freeze"]).decode()
    for line in reqs.split("\n"):
        if any(pkg in line.lower() for pkg in ["tensorflow", "cuda", "cudnn"]):
            print(f"  {line}")

check_current_state()
```

```python
# Second cell modifications
import os

import tensorflow as tf


print("\nTensorFlow Configuration:")
print(f"TensorFlow version: {tf.__version__}")

# Check CUDA environment variables
print("\nCUDA Environment:")
print(f"CUDA_VISIBLE_DEVICES: {os.getenv('CUDA_VISIBLE_DEVICES', 'Not Set')}")
print(f"LD_LIBRARY_PATH: {os.getenv('LD_LIBRARY_PATH', 'Not Set')}")

# Detailed GPU information
print("\nGPU Devices:")
physical_devices = tf.config.list_physical_devices()
print("All physical devices:", physical_devices)
print("\nGPU devices:", tf.config.list_physical_devices("GPU"))

# Check if TensorFlow can see CUDA
print("\nCUDA Available:", tf.test.is_built_with_cuda())
print("GPU Available:", tf.test.is_built_with_gpu_support())

# Try to get more detailed GPU information
if tf.test.is_built_with_cuda():
    try:
        with tf.device("/GPU:0"):
            print("\nGPU Device Details:")
            print(tf.config.get_visible_devices())
    except RuntimeError as e:
        print("\nError accessing GPU:", str(e))

# Additional NVIDIA specific information
try:
    import subprocess
    nvidia_smi = subprocess.check_output(["nvidia-smi"]).decode("utf-8")
    print("\nNVIDIA-SMI Output:")
    print(nvidia_smi)
except Exception as e:
    print("\nCould not run nvidia-smi:", str(e))
```
