```python
import os

from dotenv import load_dotenv
from stashapi import log
from stashapi.stashapp import StashInterface


# Load the .env file
load_dotenv()

# Get the API key
api_key = os.getenv("ApiKey")

stash = StashInterface(
    {
        "scheme": "http",
        "host": "fraktal.auvilankuja.fi",
        "port": "9999",
        "logger": log,
        "ApiKey": api_key
    }
)
```

```python
# Delete all markers
for i in range(1, 17272):
    try:
        stash.destroy_scene_markers(i)
    except Exception as e:
        print("Failed to delete scene marker with id: " + str(i))
        continue
```

```python
# Delete all tags
for i in range(5000, 5022):
    try:
        stash.destroy_tag(i)
    except Exception as e:
        print("Failed to delete tag with id: " + str(i))
        continue
```
