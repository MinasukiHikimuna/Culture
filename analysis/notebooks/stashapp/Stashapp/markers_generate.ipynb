```python
scene = stash.find_scene(2701)
scene["title"]
```

```python
non_video_cut_markers = [{ "id":non_video_cut_marker["id"], "name":non_video_cut_marker["primary_tag"]["name"]} for non_video_cut_marker in scene["scene_markers"] if non_video_cut_marker["primary_tag"]["id"] != "8836"]
video_cut_markers = [{ "id":video_cut_marker["id"], "name":video_cut_marker["primary_tag"]["name"]} for video_cut_marker in scene["scene_markers"] if video_cut_marker["primary_tag"]["id"] == "8836"]
```
