```python
# Build comprehensive model dictionary with full metadata
import glob
import json


def build_model_metadata():
    models_base_path = "/Volumes/Upload/Braless Forever/Braless Forever/models"
    model_metadata = {}
    model_name_to_id = {}

    # Get all model directories
    model_dirs = glob.glob(os.path.join(models_base_path, "*"))

    for model_dir in model_dirs:
        if os.path.isdir(model_dir):
            model_id = os.path.basename(model_dir)
            json_file = os.path.join(model_dir, f"{model_id}.json")

            if os.path.exists(json_file):
                with open(json_file, "r") as f:
                    model_data = json.load(f)
                    model_metadata[model_id] = model_data

                    model_name = model_data.get("name", "")
                    if model_name:
                        model_name_to_id[model_name] = model_id

    return model_metadata, model_name_to_id

# Create the comprehensive model dictionaries
model_metadata, model_name_to_id = build_model_metadata()

print(f"Found {len(model_metadata)} models with full metadata")
print(f"Name to ID lookup table has {len(model_name_to_id)} entries")
```

```python
def build_comprehensive_model_data():
    """
    Build comprehensive model data combining:
    - Braless Forever metadata from JSON files
    - StashDB ID mappings
    - Stash app performer data (enforcing single performer per model)
    """
    comprehensive_data = {}

    # Get Braless Forever metadata
    print("Loading Braless Forever metadata...")
    braless_metadata, braless_name_to_id = build_model_metadata()
    print(f"Loaded {len(braless_metadata)} models from Braless Forever")

    # Process each model
    for braless_id, braless_data in braless_metadata.items():
        model_entry = {
            "braless_forever": {
                "id": braless_id,
                "metadata": braless_data
            },
            "stashdb": {
                "id": lookup_table.get(braless_id),
                "metadata": None
            },
            "stash_app": {
                "id": None,
                "performer": None
            }
        }

        all_stash_performers = []
        model_name = braless_data.get("name", "")

        # Get StashDB ID if available and find performers with this StashDB ID
        stashdb_id = lookup_table.get(braless_id)
        if stashdb_id:
            model_entry["stashdb"]["id"] = stashdb_id

            try:
                stash_performers = stash.find_performers({
                    "stash_id_endpoint": {
                        "endpoint": "https://stashdb.org/graphql",
                        "modifier": "EQUALS",
                        "stash_id": stashdb_id
                    }
                }, fragment="id name stash_ids { stash_id endpoint }")

                if stash_performers:
                    all_stash_performers.extend(stash_performers)
            except Exception as e:
                print(f"Warning: Could not fetch Stash data for {model_name}: {e}")

        # Also try to find by Braless Forever name
        if model_name:
            try:
                braless_tagged_name = f"Braless Forever: {model_name}"
                braless_performers = stash.find_performers({
                    "name": {
                        "value": braless_tagged_name,
                        "modifier": "EQUALS"
                    }
                }, fragment="id name stash_ids { stash_id endpoint }")

                if braless_performers:
                    # Add performers not already found by StashDB ID
                    existing_ids = [p["id"] for p in all_stash_performers]
                    for performer in braless_performers:
                        if performer["id"] not in existing_ids:
                            all_stash_performers.append(performer)
            except Exception as e:
                print(f"Warning: Could not fetch Braless Forever tagged performer for {model_name}: {e}")

        # Enforce single performer constraint
        if len(all_stash_performers) > 1:
            performer_names = [p["name"] for p in all_stash_performers]
            performer_ids = [p["id"] for p in all_stash_performers]
            raise ValueError(f"Multiple Stash performers found for {model_name} (Braless ID: {braless_id}): {performer_names} (IDs: {performer_ids}). Each model should have exactly one Stash performer.")
        elif len(all_stash_performers) == 1:
            model_entry["stash_app"]["id"] = all_stash_performers[0]["id"]
            model_entry["stash_app"]["performer"] = all_stash_performers[0]

        comprehensive_data[braless_id] = model_entry

    return comprehensive_data

# Build the comprehensive data structure
print("Building comprehensive model data structure...")
comprehensive_models = build_comprehensive_model_data()
print(f"Built comprehensive data for {len(comprehensive_models)} models")

# Create quick lookup dictionaries
models_by_name = {}
models_by_stashdb_id = {}

for braless_id, model_data in comprehensive_models.items():
    # Index by name
    model_name = model_data["braless_forever"]["metadata"].get("name")
    if model_name:
        models_by_name[model_name] = model_data

    # Index by StashDB ID
    stashdb_id = model_data["stashdb"]["id"]
    if stashdb_id:
        models_by_stashdb_id[stashdb_id] = model_data

print(f"Created name lookup for {len(models_by_name)} models")
print(f"Created StashDB ID lookup for {len(models_by_stashdb_id)} models")
```

```python
def get_model_by_name(name):
    """Get comprehensive model data by Braless Forever name"""
    return models_by_name.get(name)

def get_model_by_stashdb_id(stashdb_id):
    """Get comprehensive model data by StashDB ID"""
    return models_by_stashdb_id.get(stashdb_id)

def get_model_by_braless_id(braless_id):
    """Get comprehensive model data by Braless Forever ID"""
    return comprehensive_models.get(braless_id)

def print_model_summary(model_data):
    """Print a nice summary of model data"""
    if not model_data:
        print("Model not found")
        return

    bf_data = model_data["braless_forever"]["metadata"]
    print(f"=== {bf_data.get('name', 'Unknown')} ===")
    print(f"Braless Forever ID: {model_data['braless_forever']['id']}")
    print(f"StashDB ID: {model_data['stashdb']['id'] or 'None'}")
    print(f"Stash App ID: {model_data['stash_app']['id'] or 'None'}")

    if model_data["stash_app"]["performer"]:
        performer = model_data["stash_app"]["performer"]
        print(f"Stash Performer: {performer['name']} (ID: {performer['id']})")

    # Show some key metadata
    if bf_data.get("aliases"):
        print(f"Aliases: {', '.join(bf_data['aliases'])}")
    if bf_data.get("height"):
        print(f"Height: {bf_data['height']}")
    if bf_data.get("weight"):
        print(f"Weight: {bf_data['weight']}")
    if bf_data.get("measurements"):
        print(f"Measurements: {bf_data['measurements']}")

    print()

# Test the data structure with a few examples
print("=== Testing Data Structure ===")

# Test by name lookup
rocky_data = get_model_by_name("Rocky")
if rocky_data:
    print_model_summary(rocky_data)

# Test by StashDB ID
ellie_luna_data = get_model_by_stashdb_id("226c4056-38ea-46ac-aff0-0f9cb7a160db")
if ellie_luna_data:
    print_model_summary(ellie_luna_data)

# Show stats
print(f"=== Statistics ===")
models_with_stashdb = sum(1 for m in comprehensive_models.values() if m["stashdb"]["id"])
models_with_stash_performer = sum(1 for m in comprehensive_models.values() if m["stash_app"]["id"])

print(f"Total models: {len(comprehensive_models)}")
print(f"Models with StashDB mapping: {models_with_stashdb}")
print(f"Models with Stash app performer: {models_with_stash_performer}")
```
