```python
import os
import pandas as pd
from urllib.parse import urlparse

# Assuming df is your DataFrame and it has columns 'url' and 'name'
commands = []
for index, row in df.iterrows():
    url = row['url']
    parsed_url = urlparse(url)
    old_filename = os.path.basename(parsed_url.path)
    new_filename = f"Bratty Bunny - {row['name']} (STOLEN).mp4"
    command = f'Rename-Item -Path "{old_filename}" -NewName "{new_filename}"'
    commands.append(command)

df_commands = pd.DataFrame(commands, columns=['command'])
```

```python
df_commands['command'].to_csv('rename_commands.txt', index=False, header=False, sep='\t')
```
