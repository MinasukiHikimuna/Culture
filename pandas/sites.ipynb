```python
import pandas as pds

from decouple import Config, RepositoryEnv
from sqlalchemy import create_engine

DOTENV_FILE = '/Users/thardas/Git/CultureExtractor/.env'
config = Config(RepositoryEnv(DOTENV_FILE))

host = config("HOST")
port = config("PORT")
database = config("DATABASE")
username = config("USERNAME")
password = config("PASSWORD")

connection_string = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/cultureextractor'

alchemyEngine   = create_engine(connection_string, pool_recycle=3600)
dbConnection    = alchemyEngine.connect()
```

```python
dataFrame = pds.read_sql("select * from \"sites\"", dbConnection)
pds.set_option('display.expand_frame_repr', False)
print(dataFrame)
```
