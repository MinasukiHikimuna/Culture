```python
import pandas as pd
import Levenshtein
import os
from datetime import datetime
import re

def levenshtein_ratio(name1, name2):
    return Levenshtein.ratio(name1, name2)

def parse_date_from_filename(filename):
    # Regular expression pattern for YYYY-mm-DD
    pattern = r'\d{4}-\d{2}-\d{2}'
    
    # Search for the pattern in the filename
    match = re.search(pattern, filename)
    
    # If a match is found, parse it into a date
    if match:
        date_str = match.group()
        date_object = datetime.strptime(date_str, '%Y-%m-%d')
        return date_object.date()  # Use .date() to get just the date part
    
    # If no date found, return None or handle as needed
    return None

def parse_code_from_filename(filename):
    # Split the filename by " - "
    sections = filename.split(" - ")
    
    # If there are at least three sections, return the third one
    if len(sections) >= 3:
        return sections[2]
    
    # If there are not enough sections, return None or handle as needed
    return None

def get_files_from_directory_as_df(directory_path):
    file_data = [(os.path.join(directory_path, f), f) for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
    df = pd.DataFrame(file_data, columns=["full_path", "filename"])
    return df

# Set a threshold for matching
threshold = 0.80  # You can adjust this value
```

```python
scenes_path = "Y:\\Culture\\Videos\\Sites\\Vixen Network - Blacked\\"
df_scenes = get_files_from_directory_as_df(scenes_path)

df_scenes["code"] = df_scenes["filename"].apply(parse_code_from_filename)
df_scenes["release_date"] = df_scenes["filename"].apply(parse_date_from_filename)
```

```python
galleries_path = "Z:\\Blacked Galleries\\"
df_galleries = get_files_from_directory_as_df(galleries_path)

df_galleries["code"] = df_galleries["filename"].apply(parse_code_from_filename)
df_galleries["release_date"] = df_galleries["filename"].apply(parse_date_from_filename)
```

```python
df_common = df_galleries.merge(df_scenes, on='code', how='inner')
```

```python
import shutil

target_directory = "Z:\\Blacked Galleries\\SELECTED\\"

# Iterate over each file path in the 'full_path_x' column
for file_path in df_common['full_path_x']:
    # Get the file name from the file path
    file_name = os.path.basename(file_path)
    
    # Construct the destination path in the target directory
    destination_path = os.path.join(target_directory, file_name)
    
    # Copy the file to the target directory
    shutil.move(file_path, destination_path)
```
