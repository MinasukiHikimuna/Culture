```python
import polars as pl
import sys
import os

sys.path.append(os.path.dirname(os.path.abspath('')))

from libraries.client_stashapp import get_stashapp_client, StashAppClient

stash = get_stashapp_client()
stash_client = StashAppClient()
```

```python
stash_tags = stash_client.get_tags_by_names(["Group Makeup Calculated"])
```

```python
group_makeup_calculated_tags = stash.find_tags(
    {
        "parents": {
            "value": [stash_tags.group_makeup_calculated["id"]],
            "modifier": "INCLUDES",
            "depth": 0,
            "excludes": [],
        }
    },
    fragment="id name",
)
group_makeup_calculated_tags
```

```python
stash_scenes_with_group_makeup_calculated = stash.find_scenes(
    {
        "tags": {
            "value": [stash_tags.group_makeup_calculated["id"]],
            "modifier": "INCLUDES",
            "depth": -1,
            "excludes": [],
        }
    },
    fragment="id title tags { id name }",
)
stash_scenes_with_group_makeup_calculated
```

```python
tag_ids = [tag["id"] for tag in stash_tags.group_makeup_calculated["children"]]
scene_ids = [scene["id"] for scene in stash_scenes_with_group_makeup_calculated]

stash_client.bulk_scene_update(scene_ids, tag_ids, "REMOVE")
```
